```python
decoders = {
    "CCT Fdrop" : [0.578000009059906, 0.5809999704360962, 0.5809999704360962],
    "CCT Fnoise": [0.5509999990463257, 0.5849999785423279, 0.5860000252723694],
    "CCT Dropout": [0.5860000252723694, 0.5809999704360962, 0.5899999737739563],
    "CCT G-Cutout": [0.5820000171661377, 0.5830000042915344, 0.5860000252723694],
    "CCT I-VAT": [0.52, 0.54, 0.53], # Fake
    "CCT Obj-Msk": [0.51, 0.53, 0.54], # Fake
    "CCT Cons-Msk": [0.50, 0.55, 0.56], # Fake
    "CCT full": [0.51, 0.50, 0.49], # Fake
    "CCT full+abcE": [0.56, 0.55, 0.52], # Fake
}

style = {
    "CCT Fdrop" : "o",
    "CCT Fnoise": "s",
    "CCT Dropout": "^",
    "CCT G-Cutout": "p",
    "CCT I-VAT": "*",
    "CCT Obj-Msk": "<",
    "CCT Cons-Msk": ">",
    "CCT full": "h",
    "CCT full+abcE": "H",
}

color = {
    "CCT Fdrop" : "#00b945",
    "CCT Fnoise": "#ff9500",
    "CCT Dropout": "#ff2c00",
    "CCT G-Cutout": "#845b97",
    "CCT I-VAT": "#474747",
    "CCT Obj-Msk": "#9e9e9e",
    "CCT Cons-Msk": "#1f77b4",
    "CCT full": "#00ffff",
    "CCT full+abcE": "#d2b48c"
}
```

```python
plt.rcParams.update({'font.size': 20})
fig, ax = plt.subplots()
x = np.array([2,4,6])
plt.figure(figsize=(12,8))
lstLines = []
lstLabels = []
for k, v in decoders.items():
    y = np.array([100*i for i in v])
    line,  = ax.plot(x, y, color[k], marker=style[k], markersize=14, label=k)
    lstLines.append(line)
    lstLabels.append(k)
ax.set_xticks(x)
ax.set_yticks(np.arange(35, 70, step=5))
ax.set_title("n = 1000 (10%)")
ax.set_xlabel("Number of Aux. Decoders K")
ax.set_ylabel("mIoU")
ax.legend(lstLines, lstLabels, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
