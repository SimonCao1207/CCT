```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]

from itertools import cycle

color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
```

```python
decoders = {
    "CCT Fnoise": [0.5509999990463257, 0.5849999785423279, 0.5860000252723694],
    "CCT Dropout": [0.5860000252723694, 0.5809999704360962, 0.5899999737739563],
    "CCT G-Cutout": [0.5820000171661377, 0.5830000042915344, 0.5860000252723694],
    # Fake
    "CCT I-VAT": [0.578000009059906, 0.5809999704360962, 0.5809999704360962],
    "CCT Obj-Msk": [0.578000009059906, 0.5809999704360962, 0.5809999704360962],
    "CCT Cons-Msk": [0.578000009059906, 0.5809999704360962, 0.5809999704360962],
    "CCT full": [0.578000009059906, 0.5809999704360962, 0.5809999704360962],
    "CCT full+abcE": [0.578000009059906, 0.5809999704360962, 0.5809999704360962],
}

style = {
    "CCT Fnoise": "s",
    "CCT Dropout": "^",
    "CCT G-Cutout": "p",
    "CCT I-VAT": "*",
    "CCT Obj-Msk": "<",
    "CCT Cons-Msk": ">",
    "CCT full": "h",
    "CCT full+abcE": "H",
}
```

```python
x = np.array([2,4,6])
plt.figure(figsize=(8,10))
for k, v in decoders.items():
    # print(v)
    y = np.array([100*i for i in v])
    plt.plot(x, y, next(color_cycle), marker=style[k], markersize=12)
plt.xticks(x)
plt.yticks(np.arange(35, 70, step=5))
plt.xlabel("Number of Aux. Decoders K")
plt.ylabel("mIoU")
plt.show()
```
